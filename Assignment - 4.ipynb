```python
# Now lets perform the label encoding on the all varaibles into numerical
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

dataset.Attrition = le.fit_transform(dataset.Attrition)

dataset.BusinessTravel = le.fit_transform(dataset.BusinessTravel)
dataset.Gender = le.fit_transform(dataset.Gender)
dataset.JobRole = le.fit_transform(dataset.JobRole)

dataset.OverTime = le.fit_transform(dataset.OverTime)
dataset.Department  = le.fit_transform(dataset.Department)
dataset.MaritalStatus = le.fit_transform(dataset.MaritalStatus)
dataset.EducationField  = le.fit_transform(dataset.EducationField)
```

```python
dataset.head()
```

```python
# Now all columns are converted into numerical variables into integer format 
#lets perform scaling to apply models onto the data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(dataset)
x_scaled_df = pd.DataFrame(x_scaled)
```

```python
x_scaled_df.head()
```

```python
#Now let's split the pure dependent and independent variables
x = x_scaled_df.drop(labels = [1], axis = 1)
x.head()
```

```python
y = x_scaled_df.iloc[:,1:2]
y.head()
```

```python
#train_test_split funtion
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Model Building

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_curve
import joblib  # For model saving
```

```python
logistic_model = LogisticRegression(random_state=1)
decision_tree_model = DecisionTreeClassifier(random_state=1)
random_forest_model = RandomForestClassifier(random_state=1)
```

```python
# Training and testing the Logistic Regression model
logistic_model.fit(x_train, y_train)
logistic_predictions = logistic_model.predict(x_test)

# Training and testing the Decision Tree model
decision_tree_model.fit(x_train, y_train)
decision_tree_predictions = decision_tree_model.predict(x_test)

# Training and testing the Random Forest model
random_forest_model.fit(x_train, y_train)
random_forest_predictions = random_forest_model.predict(x_test)
```

```python
logistic_accuracy = accuracy_score(y_test, logistic_predictions)
logistic_report = classification_report(y_test, logistic_predictions)
log = accuracy_score(y_train, logistic_model.predict(x_train))
print("Logistic Regression Model Accuracy for train data:", log)
print("Logistic Regression Model Accuracy:", logistic_accuracy)
print("Logistic Regression Model Classification Report:")
print(logistic_report)
```

```python
# Evaluation of Decision Tree model
decision_tree_accuracy = accuracy_score(y_test, decision_tree_predictions)
decision_tree_report = classification_report(y_test, decision_tree_predictions)

print("Decision Tree Model Accuracy:", decision_tree_accuracy)
print("Decision Tree Model Classification Report:")
print(decision_tree_report)
```

```python
# Evaluation of Random Forest model
random_forest_accuracy = accuracy_score(y_test, random_forest_predictions)
random_forest_report = classification_report(y_test, random_forest_predictions)

print("Random Forest Model Accuracy:", random_forest_accuracy)
print("Random Forest Model Classification Report:")
print(random_forest_report)
```
